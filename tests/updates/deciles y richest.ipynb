```python
import enzope
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = enzope.CPUModel(n_agents=100, f=.3, measure_every=1)
model.run(5000)
```

```python
top10 = enzope.measures.top10(model.deciles)
bottom50 = enzope.measures.bottom50(model.deciles)
middle40 = enzope.measures.middle40(model.deciles)
```

```python
plt.plot(top10, label='Top 10%')
plt.plot(bottom50, label='Bottom 50%')
plt.plot(middle40, label='Middle 40%')
```
