```python
import enzope
import numpy as np
import matplotlib.pyplot as plt
```

```python
lamb = 1
sizes = np.logspace(2, 5, 100, dtype=int)
ginis = []
for size in sizes:
    w = np.random.exponential(lamb, size)
    ginis.append(enzope.gini(w))
plt.xlabel('Sample size')
plt.ylabel('Gini coefficient')
plt.scatter(sizes, ginis)
```
